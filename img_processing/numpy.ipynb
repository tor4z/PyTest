```python
import numpy as np

image_shape = (30, 30, 3)
square_size = 5

grid = np.ogrid[[slice(i) for i in image_shape]]
grid = [(grid_i // square_size) & 1 for grid_i in grid]

checkerboard = np.bitwise_xor.reduce(grid, axis=0)
res = np.int8(checkerboard)

print(grid)
print(res)
```

```python
import numpy as np
import matplotlib.pyplot as plt

step = 0.2

x = np.arange(-10,10,step)
y = np.arange(-10,10,step)

X,Y = np.meshgrid(x,y)
print(X)
print(Y)
Z = X**2+Y**2
#plt.contourf(X,Y,Z)
plt.contour(X,Y,Z)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

image_shape = (30, 30)
square_size = 5

grid = np.ogrid[[slice(i) for i in image_shape]]
grid = [(grid_i // square_size) & 1 for grid_i in grid]

checkerboard = np.bitwise_xor.reduce(grid, axis=0)
res = np.int8(checkerboard)

plt.contour(res)
plt.show()
```
