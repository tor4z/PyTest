```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('./ss.png')
img = img[:,:,::-1]/255.0
plt.imshow(img)
plt.show()
```

```python
img_g = np.gradient(img, edge_order=2)
plt.subplot(1,3,1)
plt.imshow(img_g[0])
plt.subplot(1,3,2)
plt.imshow(img_g[1])
plt.subplot(1,3,3)
plt.imshow(img_g[2])
plt.show()
```

```python
out = 0
for i in img_g:
    if isinstance(out, int):
        out = i
    else:
        out *= i;
plt.imshow(out)
plt.show()
```

```python
unit = np.float32(np.int8(img>0.6))
print(unit.shape)
plt.imshow(unit[:,:,0])
plt.show()
plt.imshow(unit[:,:,1])
plt.show()
plt.imshow(unit[:,:,2])
plt.show()

unit_r = unit[:,:,0]
```

```python
layers = []
for i in range(3):
    layer = img[:,:,i]
    layer = layer * unit_r
    layers.append(layer)
    
    plt.imshow(layer)
    plt.show()
    
img = np.stack(layers, 2)

plt.imshow(img)
plt.show()
```
